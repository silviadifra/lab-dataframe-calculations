## How to label customers whose aggregated `amount_spent` is in a given quantile range?


We break down the main problem into several sub problems:

Sub Problem 1: How to aggregate the amount_spent for unique customers?

```python
orders.groupby(['customerid']).agg({'amount_spent':sum})
```

Sub Problem 2: How to select customers whose aggregated amount_spent is in a given quantile range?

```python
orders_groupby = orders.groupby(['customerid', 'country']).agg({'amount_spent':sum})
```

```python
orders_groupby.head()
```

```python
q25 = np.quantile(orders_groupby.amount_spent, q=.25)
q50 = np.quantile(orders_groupby.amount_spent, q=.50)
q75 = np.quantile(orders_groupby.amount_spent, q=.75)
q95 = np.quantile(orders_groupby.amount_spent, q=.95)

def quantiles_range(value):
    if value < q25:
        return "q25"
    elif value < q50:
        return "q50"
    elif value < q75:
        return "q75"
    elif value > q75 and value<q95:
        return "Preferred_Customers"
    else:
        return "VIP_Customer" 
```

```python
orders_groupby['quantiles'] = orders_groupby.amount_spent.apply(quantiles_range)
```

```python
orders_groupby.head()
```

Sub Problem 3: How to label selected customers as "VIP" or "Preferred"?

```python
VIP_Preferred_Customers = orders_groupby[(orders_groupby['quantiles']=='VIP_Customer')|(orders_groupby['quantiles']=='Preferred_Customers')] 
```

```python
VIP_Preferred_Customers
```

## Q2: How to identify which country has the most VIP Customers?

```python
# your code here
VIP_Customers= VIP_Preferred_Customers[(VIP_Preferred_Customers['quantiles'] == 'VIP_Customer')]
VIP_Customers
```

```python
most_VIP = VIP_Customers.groupby(['country', 'quantiles']).agg({'amount_spent':pd.Series.nunique}).reset_index()
most_VIP.head()
```

```python
most_VIP_final= most_VIP.sort_values(by='amount_spent', ascending=False)
most_VIP_final.head()
```

```python
most_VIP_final[['country']].loc[16]
```

```python
most_VIP_final.info()
```

```python
print('The country that has the most VIP Customers is :', most_VIP_final[['country']].loc[16])
```

## Q3: How to identify which country has the most VIP+Preferred Customers combined?

```python
# your code here
VIP_Preferred_Customers
```

```python
VIPandPrefCust_Countries=VIP_Preferred_Customers.copy()
VIPandPrefCust_Countries=VIP_Preferred_Customers.rename(columns={'quantiles':'VIP_PrefCust'})
display(VIPandPrefCust_Countries.head())
```

```python
VIPandPrefCust_Countries.reset_index()
```

```python
def count_VIPPCust(x):
    if x =="VIP_Customer" or x=='Preferred_Customers':
        return 1
```

```python
VIPandPrefCust_Countries['Most_VIP_PCust']=VIPandPrefCust_Countries.VIP_PrefCust.apply(count_VIPPCust)
display(VIPandPrefCust_Countries)
```

```python
VIPandPrefCust_Countries=VIPandPrefCust_Countries.groupby(['country']).agg({'Most_VIP_PCust':sum})
display(VIPandPrefCust_Countries.head())
```

```python
VIPandPrefCust_Countries_final=VIPandPrefCust_Countries[VIPandPrefCust_Countries['Most_VIP_PCust']==VIPandPrefCust_Countries['Most_VIP_PCust'].max()]
VIPandPrefCust_Countries_final.head()
```

```python
VIPandPrefCust_Countries_final.reset_index()
```

```python
VIPandPrefCust_Countries_final.info()
```

```python
print('The country that has the most VIP+Pref Customers is :',VIPandPrefCust_Countries['Most_VIP_PCust'].max())
```
